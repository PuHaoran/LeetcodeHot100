``` 伪代码
def quick_sort(l, r):
    if l >= r:
        return
    i, j = l-1, r+1
    x = nums[l]
    while i < j:
        while 1:
            i += 1
            if nums[i] >= x:
                break
        while 1:
            j -= 1
            if nums[j] <= x:
                break
        if i < j:
            nums[i], nums[j] = nums[j], nums[i]
    quick_sort(l, j)
    quick_sort(j+1, r)

```

|                         Alg                         |  思想         |  时间复杂度 | Tag  |
| :----------------------------------------------------------: |:-------------------: |:-------------------: |:-------------------: |
| **1. 两数之和** | 遍历数组，若当前元素存在则直接返回结果，否则用字典k:v保存target-当前元素:索引。 | O(n) | 字典 数组 |
| **4. 寻找两个正序数组的中位数** | 分治解法。将原问题转化为求两数组第k小的数，然后分别考虑两数组长度和奇偶情况，奇数取中间值，偶数取中间两数均值。重点是分治法求两数组第k小的数，因为是排过序的，若A的k//2元素小于B的k//2元素，则A的前半部分可以舍弃，原问题变为求k-(k//2)小的数；否则舍弃B的前半部分。当A数组为空或k=1时递归终止。 | O(logn) | 分治 |
| **-** | - | - | - |
| **-** | - | - | - |

##### 字符串

|                         Alg                         |  思想         |  时间复杂度 | Tag  |
| :----------------------------------------------------------: |:-------------------: |:-------------------: |:-------------------: |
| **5. 最长回文子串** | 当前(i,i)和(i,i+1)节点向两边扩散，找到对称的子串。 | $n^{2}$ | 字符串 |
| **-** | - | - | - |
| **-** | - | - | - |
| **-** | - | - | - |
| **-** | - | - | - |

##### 二分

|                         Alg                         |  思想         |  时间复杂度 | Tag  |
| :----------------------------------------------------------: |:-------------------: |:-------------------: |:-------------------: |
| **-** | - | - | - |
| **-** | - | - | - |
| **-** | - | - | - |
| **-** | - | - | - |
| **-** | - | - | - |

##### 链表

|                         Alg                         |  思想         |  时间复杂度 | Tag  |
| :----------------------------------------------------------: |:-------------------: |:-------------------: |:-------------------: |
| **2. 两数相加** | 一直接遍历，先遍历两链表，然后遍历剩余链表。建立一个头节点作为新建链表的起始节点（哨兵节点）。二递归解题，递归是当前两个节点和上一个进位相加作为新建节点的值。 | O(n) | - |
| **-** | - | - | - |
| **-** | - | - | - |
| **-** | - | - | - |
| **-** | - | - | - |
| **-** | - | - | - |

##### 栈/队列/堆

``` 伪代码
// 堆排序模版
// 建堆。构建1-n数组。
global n
arr = arr[1...n]
n = len(arr)
// 重构堆。从倒数第一个非叶子节点从后向前down(i)，以递归形式对节点位置进行调整。
def down(i):
    t = i
    if 2*i<=n and arr[2*i] < arr[t]:
        t = 2*i
    if 2*i+1<=n and arr[2*i+1] < arr[t]:
        t = 2*i+1
    if i != t:
        arr[i], arr[t] = arr[t], arr[i]
        down(t)
for i in range(n//2, 0, -1):
    down(i)
// 去掉堆顶元素
arr[1], arr[n] = arr[n], arr[1]
n -= 1
down(1)
```
